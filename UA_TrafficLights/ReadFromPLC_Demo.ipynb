# Read values from PLC demo

```python
import logging
from asyncua import Client

async def print_nested_children(node, depth=0):
    try:
        children = await node.get_children()
        for child in children:
            browse_name = await child.read_browse_name()
            print("  " * depth + str(browse_name))
            await print_nested_children(child, depth + 1)
    except Exception as e:
        print("Error:", e)

async def main():
    try:
        logging.getLogger('asyncua').setLevel(logging.WARNING)  # Suppress INFO messages
        async with Client(url='opc.tcp://10.195.1.52:4842') as client:
            root = client.nodes.root
            await print_nested_children(root)
    except Exception as e:
        print("Error connecting to the server:", e)

# Run the main function
await main()
```
